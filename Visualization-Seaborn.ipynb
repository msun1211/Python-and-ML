## Seaborn is a Python data statistical visulization based on matplotlib 
### github repository https://github.com/mwaskom/seaborn
### tutorial https://seaborn.pydata.org/tutorial.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
from statsmodels import robust
```

```python
tips = sns.load_dataset('tips')
#seaborn.load_dataset https://github.com/mwaskom/seaborn-data
```

```python
tips.head(5)
```

```python
##distplot
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
#categorical plots
```

```python
sns.barplot(x='sex', y='total_bill',data=tips)
```

```python
sns.barplot(x='day', y='total_bill', hue="sex", data=tips)
```

```python
from numpy import median
sns.barplot(x='day', y='total_bill',  data=tips, estimator=median)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='sex',data=tips, palette='Blues_d')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
##heatmap
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

## Pairplot

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette='rainbow')
```

```python
g= sns.FacetGrid(tips, col='time', row='smoker')
g=g.map(plt.hist, 'total_bill')
```

```python
g= sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
g=g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
## jointplot
g=sns.JointGrid(x='total_bill', y='tip', data=tips)
g=g.plot(sns.regplot, sns.distplot)
```

## Seaborn Linear plots

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='size', y='tip', data=tips);
```

```python
import numpy as np
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

```python
anscombe = sns.load_dataset("anscombe")
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"));
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"), order=2);
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"));
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"), robust=True, ci=None);
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"));
```

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
              scatter_kws={"s": 80});
```

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
              scatter_kws={"s": 80});
```
